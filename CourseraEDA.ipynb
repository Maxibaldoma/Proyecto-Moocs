```python
import pandas as pd
dfcc = pd.read_csv("Coursera_courses.csv")
dfcc.head()
```

```python
dfcr = pd.read_csv("Coursera_reviews.csv")
dfcr.head()
```

Comienzo cargando los datos y observo que se pueden unir estos dataframe.

```python
#unir los df
dfc = dfcr.merge(dfcc, on='course_id', how='left')
dfc
```

```python
dfc.describe().round(3)
```

Aqui podemos observar que el promedio de rating es muy alto, lo que nos indica que la calidad de los cursos que se ofrecen
cuentan de una buena calidad en cuanto al contenido.

```python
dfc.info()
```

```python
dfc.isnull().sum()
```

```python
#Cambiaria en la columna date_reviews el tipo de dato de str a datatime
dfc['date_reviews'] = pd.to_datetime(dfc['date_reviews'])
```

Cambio el tipo de dato en la columna date_reviews ya que se trata de fecha.

```python
# Contar filas duplicadas en el DataFrame
num_duplicates = dfc.duplicated().sum()

print("Cantidad de filas duplicadas:", num_duplicates)
```

```python
# Eliminar las filas duplicadas del DataFrame
dfcnuevo = dfc.drop_duplicates()
dfcnuevo.head(3)
```

Se eliminan todas aquellas filas duplicadas.

```python
# cantidad de ventas para cada nivel de rating en los cursos
ventas_por_rating = dfcnuevo.groupby('rating')['course_id'].count()

print(ventas_por_rating)
```

Ventas de acuerdo al nivel de rating

```python
import matplotlib.pyplot as plt
# Datos de ventas por rating
ventas_por_rating = dfcnuevo.groupby('rating')['course_id'].count()

# Crear el grÃ¡fico de barras
plt.bar(ventas_por_rating.index, ventas_por_rating.values)

# Agregar etiquetas y tÃ­tulo
plt.xlabel('Rating')
plt.ylabel('Cantidad de ventas')
plt.title('Ventas por rating en los cursos')

plt.show()
```

Aqui como se observa los cursos que mas demanda tienen, cuentan con un rating alto. 

```python
# Agrupar los datos por aÃ±o y calcular la media del rating
ratings_por_aÃ±o = dfcnuevo.groupby(dfcnuevo['date_reviews'].dt.year)['rating'].mean()

# Visualizar los resultados en un grÃ¡fico de lÃ­nea
plt.plot(ratings_por_aÃ±o.index, ratings_por_aÃ±o.values)
plt.xlabel('AÃ±o')
plt.ylabel('Rating promedio')
plt.title('Tendencia temporal del rating')
plt.show()
```

Se puede ver que la tendencia temporal del promedio de rating siempre fue buena aunque despues del aÃ±o 2019 mejorÃ³ aun mÃ¡s.
